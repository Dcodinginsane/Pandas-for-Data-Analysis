###  2. Distribution Plots
* used for univariate analysis
* used to find out the distribution
* Range of the observation
* Central Tendency
* is the data bimodal?
* Are there outliers?

Plots under distribution plot

* histplot
* kdeplot
* rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# ploting univariate histogram
sns.histplot(data = tips , x='total_bill')
```

```python
sns.displot(data = tips , x = 'total_bill' , kind = 'hist')
```

```python
# bins parameters
sns.displot(data =tips , x='total_bill' , kind='hist' , bins =2)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data =tips , x ='day' , kind='hist')
```

```python
# hue parameters

sns.displot(data = tips , x='tip' , kind='hist' , hue='sex')
```

```python

# element  - step
sns.displot(data = tips , x='tip' , kind='hist' , hue='sex' , element = 'step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic , x='age' , kind ='hist' , element='step')
```

```python
# faceting using col and row
sns.displot(data = tips , x = 'tip' , kind='hist' , col = 'sex' , element='step')
```

```python
#kdeplot
# Rather than using discrete bins a KDE smooths the obervations with a Gussian kernel , producing a continous density estimate
sns.kdeplot(data = tips , x = 'total_bill')
```

```python
# hue -> fill
sns.displot(data=tips , x ='total_bill' , kind='kde' , hue='sex' , fill = True)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
sns.kdeplot(data = tips , x='total_bill')
sns.rugplot(data = tips , x = 'total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')

sns.histplot(data =tips , x='total_bill' , y='tip')
```

```python
sns.displot(data = tips , x='total_bill' , y='tip' , kind ='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### Matrix Plot
* Heatmap
* Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix

temp_df= gap.pivot(index = 'country' , columns = 'year' , values='lifeExp')
temp_df
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index = 'country' , columns = 'year' , values='lifeExp')
temp_df
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df , annot=True , linewidths=0.5, cmap = 'viridis')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

### Q-4: Draw a scatter plot based on the below conditions:
1. x-axis should be "age" and y-axis should be "bmi".
2. For hue, size and style parameters use "diabetic", "gender" and "smoker" column respectively.
3. Add title to your chart.
4. Age should be less than 70 percentiles.
5. BMI should be greater than the average value of the filtered age dataset.

```python
import numpy as np
import pandas as pd
```

```python
# code here
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQVpcVtdYdZU4zU4-lqxt-iPHkyndDWs_aqEDUu9ZodlJ48Dku0PFgdXlj2N5RCmwXJrNtZLsI_wEVf/pub?gid=220677750&single=true&output=csv')
df.head()
```

```python
temp_df = df[df['age'] < df['age'].quantile(0.70)]
temp_df = temp_df[temp_df['bmi'] > temp_df['bmi'].mean()]
```

```python
plt.subplots(figsize=(12,8))
sns.scatterplot(data=temp_df, x='age', y='bmi', hue='diabetic',size='claim', style='smoker')
plt.show()
```
