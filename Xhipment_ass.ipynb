```python
import pandas as pd

# Convert the 'created_at' and 'updated_at' columns to datetime, handling potential format issues
df1['created_at1'] = pd.to_datetime(df1['created_at'], errors='coerce')  # Handle any incorrect formats
df1['updated_at1'] = pd.to_datetime(df1['updated_at'], errors='coerce')  # Handle timezone and format issues

# 1. Add Quote Accepted Date (created_at)
df1['Quote_Accepted_Date'] = df1['created_at1']

# 2. Add Booking Confirmation Date (if status is 'Booking Confirmed')
df1['Booking_Confirmation_Date'] = df1.apply(
    lambda row: row['updated_at1'] if row['status'] == 'Booking Confirmed' else pd.NaT,
    axis=1
)

# 3. Add Delivery Date (if status is 'Delivered to Consignee')
df1['Delivery_Date'] = df1.apply(
    lambda row: row['updated_at1'] if row['status'] == 'Delivered to Consignee' else pd.NaT,
    axis=1
)

# Display the DataFrame to verify changes
print(df1[['booking_id', 'status', 'Quote_Accepted_Date', 'Booking_Confirmation_Date', 'Delivery_Date']])
```

```python
import pandas as pd
import numpy as np
import json

# Step 1: Parse the delivery and booking dates from the 'milestones' field
def extract_dates(row):
    try:
        milestones = json.loads(row['milestones'])  # Use json.loads to handle JSON properly
    except (json.JSONDecodeError, TypeError):
        return None, None

    delivered_date = None
    booking_confirmed_date = None
    for milestone in milestones:
        if milestone['value'] == 'Delivered to Consignee':
            delivered_date = pd.to_datetime(milestone.get('dateTime', None))
        elif milestone['value'] in ['Quote Accepted', 'Booking Confirmed']:
            booking_confirmed_date = pd.to_datetime(milestone.get('dateTime', None))

    return delivered_date, booking_confirmed_date

# Apply the function to extract dates
df1[['delivered_date', 'booking_confirmed_date']] = df1.apply(extract_dates, axis=1, result_type='expand')

# Step 2: Calculate the delivery time
df1['delivery_time'] = (df1['delivered_date'] - df1['booking_confirmed_date']).dt.days

# Filter for records in 2023 only
df_2023 = df1[(df1['delivered_date'] >= '2023-01-01') & (df1['delivered_date'] < '2024-01-01')]

# Extract month and product type for grouping
df_2023['month'] = df_2023['delivered_date'].dt.month
df_2023['product_type'] = df_2023['kind']  # Assuming 'kind' refers to product type like FCL, LCL, etc.

# 1. Calculate average delivery time per product type per month
average_delivery_time = df_2023.groupby(['product_type', 'month'])['delivery_time'].mean().reset_index()

# 2. Calculate TP90 delivery time per product type per month
tp90_delivery_time = df_2023.groupby(['product_type', 'month'])['delivery_time'].apply(lambda x: np.percentile(x, 90)).reset_index()

# 3. Count number of customers with at least 1 delivery (assuming there's a 'customer_id' field)
df_2023['customer_id'] = df_2023['customer_invoice_no']  # Replace with actual customer ID if different
customers_with_delivery = df_2023.groupby('customer_id').size().reset_index(name='num_deliveries')
customers_with_at_least_1_delivery = customers_with_delivery[customers_with_delivery['num_deliveries'] >= 1]['customer_id'].nunique()

# Output results
print("Average Delivery Time per Product Type per Month:")
print(average_delivery_time)

print("\nTP90 Delivery Time per Product Type per Month:")
print(tp90_delivery_time)

print("\nNumber of Customers with at Least 1 Delivery:")
print(customers_with_at_least_1_delivery)


```
